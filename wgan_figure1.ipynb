# WGAN Paper Figure 1 Reproduction

This notebook reproduces the example comparing:
- Wasserstein distance (Earth Mover)
- JS divergence

Between two Dirac distributions:
- Real: $P_r = \delta(x=0)$
- Model: $P_\theta = \delta(x=\theta)$

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Generate values

```python
# theta values
theta = np.linspace(-1, 1, 100)

# Wasserstein distance (|theta|)
wasserstein = np.abs(theta)

# JS divergence
js = np.ones_like(theta) * np.log(2)
js[np.isclose(theta, 0, atol=1e-3)] = 0
```

## Plot

```python
plt.figure(figsize=(10,4))

# Wasserstein plot
plt.subplot(1,2,1)
plt.plot(theta, wasserstein, 'o', markersize=3)
plt.title("Wasserstein Distance")
plt.xlabel(r"$\theta$")
plt.ylabel(r"$W(P_r, P_\theta)$")

# JS divergence plot
plt.subplot(1,2,2)
plt.plot(theta, js, 'o', markersize=3)
plt.title("JS Divergence")
plt.xlabel(r"$\theta$")
plt.ylabel(r"$JS(P_r || P_\theta)$")

plt.tight_layout()
plt.show()
```
